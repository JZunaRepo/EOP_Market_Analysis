```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import seaborn as sns
df = pd.read_csv('data/clean_data.csv')

lin = LinearRegression()

train = df[["work_from_home"]].to_numpy().reshape(-1,1)
lin.fit(train,df["salary_standardized"])
```

```python
lin.score(train,df["salary_standardized"])
```

```python
lin.coef_
```

```python
predictions = lin.predict(train)

sns.scatterplot(data = df, x = "work_from_home", y = "salary_standardized")

plt.plot(df["work_from_home"], predictions, color='r')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(train,df["salary_standardized"])
```
