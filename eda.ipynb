```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

<h2>Full Standard Salaries Visualizations & Statistics<h2>

```python
df = pd.read_csv('data/clean_data.csv')
```

```python

sns.histplot(data=df, x ="salary_standardized", kde=True)
```

```python
sns.boxplot(y=df["salary_standardized"])
```

```python
stats.kstest(df["salary_standardized"], stats.norm.cdf, args=(df["salary_standardized"].mean(), df["salary_standardized"].std()))
```

<h2>Remote Standard Salaries Visualizations & Statistics<h2>

```python
remote_df = df[df["work_from_home"]==True]
nonremote_df = df[df["work_from_home"]==False]

sns.histplot(data=remote_df, x ="salary_standardized", kde=True)
```

```python
stats.kstest(remote_df["salary_standardized"], stats.norm.cdf, args=(remote_df["salary_standardized"].mean(), remote_df["salary_standardized"].std()))
```

<h2>Ranked Cities<h2>

```python
means = df.groupby("location").count()["salary_standardized"].reset_index()
sorted_means = means.sort_values(by="salary_standardized", ascending = False)

sns.barplot(x="salary_standardized", y="location", data=sorted_means.head(25))
```

```python
sns.barplot(x=df.location.value_counts().head(25), y=df.location.value_counts().head(25).index)
```

```python
location_list = ["Anywhere", "United States","Wichita, KS"]
locations = df[df["location"].isin(location_list)]
sns.boxplot(x=locations["location"], y=locations["salary_standardized"],showfliers= False)
```
